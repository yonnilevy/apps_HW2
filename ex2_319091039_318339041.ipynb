```python
import random
import numpy as np
import matplotlib.pyplot as plt 

# Initial conditions
N = 2000 # Population size
Max_Population = 100000 # When is "too much" bears in the colony
color_Env = 2.0 # color of the Environment
color_bear = 7.0 # Set all bears to color 7 (10 = black, 1 = white)
Variance = 0.1 # The variance in the bears colors we allow when thay reproduce
white_chance = 0.007 # chance for a white bear
max_age = 25 # we kill bears in this age
reproduce_range = (7,20) # window where bears can have children 
Simulation_time = 120 # How many generations we will simulate


Generations = range(0,Simulation_time)
Population = []
PopulationSize = []
avg_color = []
color=[]
for i in range(0,N):
    bearc = color_bear + (random.random()*2-1)*Variance*2
    bear_age = round(random.random()*70)
    Population.append([bearc,bear_age])
plot = True
# Start evolution generation by generation
for Generation in Generations:
    children = []
    color=[]
    for x in Population:
        bearcolor = x[0]
        if bearcolor - color_Env == 0:
            Survive_chance = 1
        else:
            Survive_chance = 1/np.abs(bearcolor - color_Env)
            if Survive_chance > 1:
                Survive_chance = 1
        if x[1] in range(reproduce_range[0],reproduce_range[1]) and N < Max_Population:
            childc = bearcolor + (random.random()*2-1)*Variance*2
            if childc < 1:
                childc = 1
            if random.random() <= white_chance:
                childc = color_Env
            color.append(childc)
            children.append([childc,0])
            N += 1
        if x[1] >= max_age or random.random() >= Survive_chance:
            Population.remove(x)
            N -= 1
        else:
            x[1] += 1
            color.append(bearcolor)
    if N == 0:
        plot = False
        print("all bears are dead")
        break
    Population = Population + children       
    avg_color.append(np.average(color))
    PopulationSize.append(N)

if plot:      
    print('Finished calculating the simulation, here are the results:')
    plt.plot(Generations,PopulationSize)
    plt.plot(Generations,Max_Population*np.ones(len(Generations)),"r-")
    plt.xlabel("Generation")
    plt.ylabel("Population")
    plt.show()
    plt.plot(Generations,avg_color)
    plt.plot(Generations,color_Env*np.ones(len(Generations)),"r-")
    plt.xlabel("Generation")
    plt.ylabel("bear color")
    plt.show()
```

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as npy
import random
import math

# timestep determines the accuracy of the euler method of integration
timestep = 0.0001
# amplitude of noise term
amp = 0.00
# the time at which the simulation ends
end_time = 50
# creates a time vector from 0 to end_time, seperated by a timestep
t = npy.arange(0,end_time,timestep)
# intialize rabbits (x) and foxes (y) vectors
x = []
y = []

"""" definition of lotka-volterra parameters"""
a=1 # birth rate of rabbits
b = 0.1 # death rate of rabbits due to predation
c = 0.5 # natural death rate of foxes
d = 0.02 # factor that describes how many eaten rabbits give birth to a new fox
""" calculates the change in the two population over time """

# initial conditions for the rabbit (x) and fox (y) populations at time=0
x.append(100)
y.append(20)

for index in range(1,len(t)):
    # evaluate the differentials (change in population) for this iteration
    xd = x[index-1] * (a - b*y[index-1])
    yd = -y[index-1]*(c - d*x[index-1])
    # evaluate the next value of x and y populations
    next_x = x[index-1] + xd * timestep
    next_y = y[index-1] + yd * timestep
    # add the next value of x and y
    x.append(next_x)
    y.append(next_y)
    
""" visualization """
plt.plot(t, x)
plt.plot(t, y)
plt.xlabel('Time')
plt.ylabel('Population Size')
plt.legend(('Rabbits', 'Foxes'))
plt.title('Deterministic Lotka-Volterra')
plt.show()
plt.plot(x,y)
plt.xlabel('Rabbit Population')
plt.ylabel('Fox Population')
plt.title('Phase Portrait of Deterministic Lotka-Volterra')
plt.show()
```
